```python
import pandas as pd
import scienceplots
import matplotlib.pyplot as plt

plt.style.use(['science', 'ieee'])
plt.rcParams["text.usetex"] = False
plt.rcParams['figure.figsize'] = 6, 2
```

```python
V = pd.read_feather('out_veremi/veremi.feather')

sizes = V.groupby('attack_type').size()
```

```python
f = plt.bar(sizes.index, sizes, color='black')
#plt.title('Histograma da quantidade de mensagens por ataque no conjunto de dados')
plt.xticks(sizes.index,)
plt.xlabel('Tipo de ataque')
plt.ylabel('Número de mensagens no dataset')
plt.savefig('hist-attacks.png')
```
